# Initial model

```python
# Import modules, constants
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (9, 6)  # Default plot size
import numpy as np
import function as fnc
import constants 
```

Define geometry & initial boundary conditions:

```python
# Define glacier geometry
# thk: thickness [m] 
thk = 1000
```

```python
# Define Surface Boundary Conditions
# b_dot: vertical velocity (asummed to equal snow accumulation rate) [m/yr]
# T_S: Surface Dirichlet temperature [C]
b_dot = 0.3
T_S = -30 
```

```python
# Define Basal Boundary Condition
# H_flux: Geothermal heat flux [W/m2]
H_flux = 0.05
```

Run the Steady-State model using Robin (1955) Approximation: 

```python
# Estimate the Steady-State Temperature profile
temp, depth = fnc.robin_temperature(b_dot, thk, H_flux, T_S)
```

Plot the initial temperature profile: 

```python
# Plot the results
plt.plot(temp, depth, color = 'k')  
plt.title('Initial Temperature profile')
plt.xlabel('Temperature ($^\circ$C)')
plt.ylabel('Depth (m)')
plt.show()
```

# Experiment 1 - Modify SMB (or vertical velocity)

```python
# Changes in SMB [m/yr]
decrease_b_dot = 0.1
increase_b_dot = 1
```

Run the model with the new SMB (or vertical velocities):

```python
# Estimate the Steady-State Temperature profile using Robin (1955) Approximation
decrease_new_temp, depth = fnc.robin_temperature(decrease_b_dot, thk, H_flux, T_S)
increase_new_temp, depth = fnc.robin_temperature(increase_b_dot, thk, H_flux, T_S)
```

Plot the new temperature profiles:

```python
# Plot the results
plt.plot(temp, depth, color = 'k', linestyle='dashed', label = 'Initial Temperature profile') 
plt.plot(decrease_new_temp, depth, color = 'g', label = 'Temperature profile: Decrease SMB') 
plt.plot(increase_new_temp, depth, color = 'r', label = 'Temperature profile: Increase SMB') 
plt.title('Temperature profile: Changing SMB (or vertical velocity)')
plt.xlabel('Temperature ($^\circ$C)')
plt.ylabel('Depth (m)')
plt.show()
```
